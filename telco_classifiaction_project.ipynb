```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from acquire import get_telco_data
%matplotlib inline
```

# Acquire telco data

```python
# acquire data from codeup SQL database
df = get_telco_data()
```

```python
# dataframe datatypes
df.info()
```

- column total_charges is an object (string), need to convert to float

```python
# convert total_charges column to type float
# values that cannot be converted to type float 
# will be converted to NaN
# NaN will be handled during data preperation
df['total_charges'] = pd.to_numeric(df['total_charges'],errors='coerce')
df.info()
```

```python
df.isna().sum()
```
